## 3. Regularization with Polynomial Regression as a Case Study

```python
N_REG_POINTS = 20
RANGE = 3
NOISE_COEF = 5


X = np.random.uniform(low=-RANGE, high=RANGE, size=N_REG_POINTS)
y =X**3 + NOISE_COEF*np.random.normal(size=N_REG_POINTS)

plt.scatter(X, y)
plt.show()
```

```python
POLY_DEGREE = 3
feature_generator = preprocessing.PolynomialFeatures(degree=POLY_DEGREE)
X_final = feature_generator.fit_transform(X.reshape(-1, 1))


CLIP_LIMS = 1

model = linear_model.LinearRegression()
model.fit(X_final, y)

X_to_plot = np.linspace(-RANGE, RANGE, num=100)
X_to_plot_final = feature_generator.transform(X_to_plot.reshape(-1, 1))
yhat_to_plot = model.predict(X_to_plot_final)

yhat_to_plot = np.clip(yhat_to_plot, np.amin(y)-CLIP_LIMS*(np.amax(y) - np.amin(y)), np.amax(y)+CLIP_LIMS*(np.amax(y) - np.amin(y)))



plt.scatter(X, y)
plt.plot(X_to_plot, yhat_to_plot, color="red")
plt.show()
```

## 4. Regularization in soft-SVM

```python
N_CLUSTER_POINTS = 60

MEAN_1 = [-3, 3]
COV_1 = [[2, 0.9],[0.9, 2]]

MEAN_2 = [3, -3]
COV_2 = [[2, 0.9],[0.9, 2]]

cluster_1 = np.random.multivariate_normal(mean=MEAN_1,cov=COV_1,size=N_CLUSTER_POINTS)
cluster_2 = np.random.multivariate_normal(mean=MEAN_2,cov=COV_2,size=N_CLUSTER_POINTS)
```

```python
plt.scatter(cluster_1[:,0], cluster_1[:,1])
plt.scatter(cluster_2[:,0], cluster_2[:,1], color="red")
plt.show()
```

```python
zeros = np.zeros((cluster_1.shape[0], 1))
ones = np.ones((cluster_2.shape[0],1))
cluster_1_concatted = np.concatenate([cluster_1, zeros], axis=1)
cluster_2_concatted = np.concatenate([cluster_2, ones], axis=1)

main_data = np.concatenate([cluster_1_concatted, cluster_2_concatted], axis=0)

X = main_data[:,:-1]
y = main_data[:,-1]
```

```python
svm_model = svm.LinearSVC(C=1, dual=False)
svm_model.fit(X, y)
```

```python
yhat = svm_model.predict(X)
metrics.accuracy_score(y, yhat)
```

```python
def two_d_svm_plot(model, lower_limit, upper_limit):
    coefs = model.coef_.reshape(-1) 
    intercept = model.intercept_
    X = np.linspace(lower_limit, upper_limit, num=10)
    y = (-coefs[0]*X - intercept)/coefs[1]
    plt.plot(X, y, color="black")
```

```python
plt.scatter(cluster_1[:,0], cluster_1[:,1])
plt.scatter(cluster_2[:,0], cluster_2[:,1], color="red")
two_d_svm_plot(svm_model, -6, 6)
plt.show()
```

## 5. Dealing with imbalance in labels: First Steps

```python
N_CLUSTER_POINTS_1 = 600
N_CLUSTER_POINTS_2 = 60

MEAN_1 = [-1, 1]
COV_1 = [[4, 0.25],[0.25, 4]]

MEAN_2 = [1.75, -1.75]
COV_2 = [[1, 0.25],[0.25, 1]]

cluster_1 = np.random.multivariate_normal(mean=MEAN_1,cov=COV_1,size=N_CLUSTER_POINTS_1)
cluster_2 = np.random.multivariate_normal(mean=MEAN_2,cov=COV_2,size=N_CLUSTER_POINTS_2)

plt.scatter(cluster_1[:,0], cluster_1[:,1])
plt.scatter(cluster_2[:,0], cluster_2[:,1], color="red")
plt.show()

zeros = np.zeros((cluster_1.shape[0], 1))
ones = np.ones((cluster_2.shape[0],1))
cluster_1_concatted = np.concatenate([cluster_1, zeros], axis=1)
cluster_2_concatted = np.concatenate([cluster_2, ones], axis=1)

main_data = np.concatenate([cluster_1_concatted, cluster_2_concatted], axis=0)

X = main_data[:,:-1]
y = main_data[:,-1]
```

```python
svm_model = svm.LinearSVC(C=1, dual=False)
svm_model.fit(X, y)
plt.scatter(cluster_1[:,0], cluster_1[:,1])
plt.scatter(cluster_2[:,0], cluster_2[:,1], color="red")
two_d_svm_plot(svm_model, -2, 4)
plt.show()
```

```python
svm_model = svm.LinearSVC(C=1, dual=False, class_weight={0:0.1,1:0.9})
svm_model.fit(X, y)
plt.scatter(cluster_1[:,0], cluster_1[:,1])
plt.scatter(cluster_2[:,0], cluster_2[:,1], color="red")
two_d_svm_plot(svm_model, -2, 4)
plt.show()
```
