**Matplotlib**, the basic data visualization tool of Python

```python
#Import dependencies
import numpy as np
import pandas as pd
```

```python
#Import Matplotlib
#pyplot is what we work with most of the time
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(0, 10, 100)

# create a plot figure
fig = plt.figure()

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '-');
```

The following code produces sine and cosine curves using Pyplot API.

```python
#create a plot figure
plt.figure()
```

```python
plt.subplot(2, 1, 1)   # (rows, columns, panel number)
plt.plot(x1, np.sin(x1))
```

```python
# create the second of two panels and set current axis
plt.subplot(2, 1, 2)   # (rows, columns, panel number)
plt.plot(x1, np.cos(x1));
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('Numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.show()
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, 'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title');
```

```python
#specifying both x and y lists
x3 = np.arange(0.0, 6.0, 0.01) 
plt.plot(x3, [xi**2 for xi in x3], 'b-') 
plt.show()
```

```python
x4 = range(1, 5)
plt.plot(x4, [xi*1.5 for xi in x4])
plt.plot(x4, [xi*3 for xi in x4])
plt.plot(x4, [xi/3.0 for xi in x4])
plt.show()
```
