```python
import pandas as pd
import numpy as np
```

## Creating and Manipulating Series objects

```python
a = np.array(
    np.random.randint(1, 100, 10)
)

print(a.shape)

s = pd.Series(a)
print(s)
```

```python
s = pd.Series(a, index = np.random.randint(300, 400, 10))
print(s)
```

```python
d = {
    "key1": "val1",
    "key2": "val2",
    "key3": "val3"
}

t = pd.Series(d)
print(t["key1"])
print(t[:21])
```

## Transformations, Lambda Functions, and Sorting

```python
df["SepalDoubled"] = df["SepalLengthCm"].apply(lambda x : 2.0 * x)
```

```python
df.head()
```

```python
df.sort_values(
    by = ["SepalDoubled"],
    ascending = False, 
    inplace = False
).tail(3)
```

```python
df.drop(
    ["SepalDoubled"],
    axis = "columns",
    inplace = True
)
```
